```python
import pandas as pd
#leemos el dataset
colors = pd.read_csv("https://raw.githubusercontent.com/manlio99/Materia-de-aprendizaje/master/3_MidtermProjects/ProjectRBG/data/colors.csv")
#arreglo el dataset, puesto que falta una fila
df = colors
df.loc[-1]=['air_force_blue_raf','Air Force Blue (Raf)','#5d8aa8',93,138,168]
df.index = df.index + 1
df.sort_index(inplace=True)
#renombro las columnas
df = df.rename(columns={'air_force_blue_raf':'color_name','Air Force Blue (Raf)':'color_group','#5d8aa8': 'HEX','93':'R','138':'G','168':'B'})
df
```

```python
Groups = ["Blue","Red","Pink","Green","Yellow","Orange","Brown","Purple","Grey","White","Black"]
BlueDetection = ["Blue","Sapphire","Azure","Aqua","Turquoise","Celeste","Cyan","Teal","Lapis","Cobalt"]
RedDetection = ["Red","Ruby","Scarlet","Crimson","Lava","Carmine","Vermilion","Fire","Flame"]
PinkDetection = ["Pink","Rose","Salmon","Magenta","Fuchsia","Cherry","Wild"]
GreenDetection = ["Green","Lime","Emerald","Bill","Olive","Mint","Jade"]
YellowDetection = ["Yellow","Gold","Lemon","Sand","Banana","Amber"]
OrangeDetection = ["Orange", "Tangerine","Peach","Pumpkin","Carrot","Bronze","Coral"]
BrownDetection = ["Brown","Copper","Tan","Coffee","Maroon","Chocolate"]
PurpleDetection = ["Purple","Violet","Indigo","Lavender","Amethyst","Wine","Orchid","Eggplant"]
GreyDetection = ["Grey","Gray","Silver","Charcoal","Platinum"]
WhiteDetection = ["White","Ivory","Eggshell"]
BlackDetection = ["Black","Coal","Evony","Onyx"]


for i in range(865):
  if any(x in df['color_group'][i] for x in BlueDetection) : df['color_group'][i] = 'Blue'
  if any(x in df['color_group'][i] for x in RedDetection) : df['color_group'][i] = 'Red'
  if any(x in df['color_group'][i] for x in PinkDetection) : df['color_group'][i] = 'Pink'
  if any(x in df['color_group'][i] for x in GreenDetection) : df['color_group'][i] = 'Green'
  if any(x in df['color_group'][i] for x in YellowDetection) : df['color_group'][i] = 'Yellow'
  if any(x in df['color_group'][i] for x in OrangeDetection) : df['color_group'][i] = 'Orange'
  if any(x in df['color_group'][i] for x in BrownDetection) : df['color_group'][i] = 'Brown'
  if any(x in df['color_group'][i] for x in PurpleDetection) : df['color_group'][i] = 'Purple'
  if any(x in df['color_group'][i] for x in GreyDetection) : df['color_group'][i] = 'Grey'
  if any(x in df['color_group'][i] for x in WhiteDetection) : df['color_group'][i] = 'White'
  if any(x in df['color_group'][i] for x in BlackDetection) : df['color_group'][i] = 'Black'

df =  df.sort_values(by='color_group')

UsefulDf = df[df.color_group.isin(Groups) == True]
UselessDf = df[df.color_group.isin(Groups) == False]
UsefulDf = UsefulDf.drop(["HEX","color_name"],axis=1)
UsefulDf
```

# Entrenamiento con SVM

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

X = UsefulDf.drop("color_group",axis=1) # Entradas (tablota)
Y = UsefulDf.color_group # Salidas (targets)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, shuffle=True)  # random_state = 0 para congelar el aleatorio
# Creo el modelo y lo entreno
clf = SVC(gamma=0.0001)
clf.fit(X_train, Y_train)
# Testeo (pruebo el clasificador con datos de testing)
score = clf.score(X_test, Y_test)
print('PrecisiÃ³n = ', score*100,'%')
print()
```

# Entrenamiento con KNN

```python
from sklearn.neighbors import KNeighborsClassifier
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2,shuffle=True)  # random_state = 0 para congelar el aleatorio
knn = KNeighborsClassifier(n_neighbors=100) # Como parÃ¡metro recibe la cantidad de vecinos a evaluar

# Entrenamiento
knn.fit(X_train, Y_train)

# Score/accuracy = cantidad de casos de prueba correctamente clasificados / cantidad de casos de prueba
score = knn.score(X_test, Y_test)
print('El score es ', score)
print()
```
